```python
#FAZ PARSE NOS DADOS E SALVA UM DATAFRAME
data = []
for item in resultados:
    row_data = {'link':'', 'coin_name':'', 'sigla':'', 'algorithm':'', 'preco_dolar':'', 'pools':'',
           'total_hashare':'', 'network_hashate': ''}
    
    link = item.find('a')
    if link != None:  row_data['link'] = main_url + link.get('href')
    
    coin_name = item.find('b')
    if coin_name != None:  row_data['coin_name'] = coin_name.text
    
    sigla = item.find('small')
    if sigla != None:  row_data['sigla'] = sigla.text
    
    algorithm = item.find('div',{'style':'text-align:center;padding-right:25px;padding-top:5px;'})
    if algorithm != None:  row_data['algorithm'] = algorithm.text
    
    preco_dolar = item.find('td',{'class':'show1050'})
    if preco_dolar != None:  row_data['preco_dolar'] = preco_dolar.text
    
    pools = item.find('td',{'class':'show1150'})
    if pools != None:  row_data['pools'] = pools.text
    
    total_hashare = item.find('td',{'class':'show1250'})
    if total_hashare != None:  row_data['total_hashare'] = total_hashare.text

    network_hashate = item.find('td',{'class':'show1350'})
    if network_hashate != None:  row_data['network_hashate'] = network_hashate.text
    
    data.append(row_data)
```
