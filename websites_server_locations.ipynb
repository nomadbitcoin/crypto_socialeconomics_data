```python
def getServerIP(url):
    withou_protocol = url.replace('http://', '').replace('https://', '')
    return socket.gethostbyname(withou_protocol)
```

```python
API_TOKEN = ''
MAIN_URL = 'http://ipinfo.io/'
def getWebsiteLocationServer(url, position_row):
    try:
        url_ip = getServerIP(url)
        
        url = MAIN_URL + url_ip + API_TOKEN
        response = urlopen(url)
        data = json.load(response)

        data_model = {'ip': '', 'org':'', 'city':'', 'country':'', 'region':'', 
                      'loc':'', 'postal':'', 'timezone':'', 'anycast':'', 'hostname':''}

        for key in data_model.keys():
            try:
                if data[key] != None: crypto_data.loc[position_row, key] = data[key]
            except:
                pass
            
        return True
        #EXEMPLO DE OTIMIZACAO DE LOGICA 
#         if data['ip'] != None: crypto_data.loc[position_row, 'ip'] = data['ip']
#         if data['org'] != None: crypto_data.loc[position_row, 'org'] = data['org']
#         if data['city'] != None: crypto_data.loc[position_row, 'city'] = data['city']
#         if data['country'] != None: crypto_data.loc[position_row, 'country'] = data['country']
#         if data['region'] != None: crypto_data.loc[position_row, 'region'] = data['region']
            
#         if data['loc'] != None: crypto_data.loc[position_row, 'loc'] = data['loc']
#         if data['postal'] != None: crypto_data.loc[position_row, 'postal'] = data['postal']
#         if data['timezone'] != None: crypto_data.loc[position_row, 'timezone'] = data['timezone']
#         if data['anycast'] != None: crypto_data.loc[position_row, 'anycast'] = data['anycast']
#         if data['hostname'] != None: crypto_data.loc[position_row, 'hostname'] = data['hostname']
    except:
        crypto_data.loc[position_row, 'ip_location_error'] = 'Error'
        return 'Error'
```

```python
for pos, item in crypto_data.T.items():
    if not(pos < 100): 
        getWebsiteLocationServer(item['website'], pos)
        print('{} website processed: {} and the country found is: {}'.format(pos, item['website'], crypto_data.loc[pos, 'country']))
#         if pos ==100:
#             break
```

```python
already_ploted = []
for location in crypto_data['loc']:
    try:
        if location not in already_ploted:
            location = location.split(',')
            folium.Marker(location=[location[0], location[1]], icon = folium.Icon(color='red')).add_to(loc_in_map)
            already_ploted.append(location)
    except:
        pass
loc_in_map
```
