```python
def getTelegramInfo(telegram_link):
    try:
        req = requests.get(telegram_link)
        if req.status_code == 200:
            htmlPage = bs4.BeautifulSoup(req.content, features='html.parser')
            return htmlPage.find('div', {'class':'tgme_page_extra'}).text
        else:
            return 'Error'
    except:
        return 'Error'
```

```python
def getWebsiteHtml(website_link):
    try:
        req = requests.get(website_link)
        if req.status_code == 200:
            htmlPage = bs4.BeautifulSoup(req.content, features='html.parser')

            #SAVE HTML FILE
            path_ = 'html_sites_cryptocurrencies/'+ str(website_link.replace('http://', '').replace('https://', '')) + '.txt' 
            with open(path_, 'w') as filee:
                filee.write(str(htmlPage.encode('utf8')))
            return htmlPage
        else:
            return 'Error'
    except:
        return 'Error'
```

```python
def getSocialMedia(position_dataframe, html_file):
    for link in html_file.find_all('a'):
        if link.get('href') != None and 'facebook' in link.get('href'): data.loc[position_dataframe, 'facebook'] = link.get('href')

        if link.get('href') != None and 't.me' in link.get('href'): data.loc[position_dataframe, 'telegram'] = link.get('href')

        if link.get('href') != None and 'instagram' in link.get('href'): data.loc[position_dataframe, 'instagram'] = link.get('href')

        if link.get('href') != None and 'discord' in link.get('href'): data.loc[position_dataframe, 'discord'] = link.get('href')

        if link.get('href') != None and 'twitter' in link.get('href'): data.loc[position_dataframe, 'twitter'] = link.get('href')

        if link.get('href') != None and 'youtube' in link.get('href'): data.loc[position_dataframe, 'youtube'] = link.get('href')

        if link.get('href') != None and 'bitcointalk' in link.get('href'): data.loc[position_dataframe, 'bitcointalk'] = link.get('href')

        if link.get('href') != None and 'github' in link.get('href'): data.loc[position_dataframe, 'github'] = link.get('href')

        if link.get('href') != None and 'reddit' in link.get('href'): data.loc[position_dataframe, 'reddit'] = link.get('href')
```

```python
for pos, link in data[data['Website '].isnull() == False]['Website '].T.items():
    #SE DER ALGUM ERRO NO REQUESTS IRA SALVAR O CODIGO DE RESPOSTA DO SITE
    response_ = getWebsiteHtml(link)
    if response_ == 'Error': 
        data.loc[pos, 'site_error'] = response_
    else:
        getSocialMedia(pos, response_)
        
    print('{} website processed {}'.format(pos, data.loc[pos, 'coin_name']))
```

```python
def getWebsiteHtmlBySelenium(website_link):
    try:
        driver.get(website_link)
        htmlPage = bs4.BeautifulSoup(driver.page_source, features='html.parser')
        #SAVE HTML FILE
        path_ = 'html_sites_cryptocurrencies/'+ str(website_link.replace('http://', '').replace('https://', '')) + '.txt' 
        with open(path_, 'w') as filee:
            filee.write(str(htmlPage.encode('utf8')))
        return htmlPage
    except:
        return 'Error'
```
